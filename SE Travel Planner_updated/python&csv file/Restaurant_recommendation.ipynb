```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer #tfidf
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df_reviews=pd.read_csv("reviews.csv")
```

```python
df_reviews.head()
```

```python
df_reviews['rating_category']
```

```python
tfidf=TfidfVectorizer(stop_words="english")#remove fillers
df_reviews['rating_category']=df_reviews['rating_category'].fillna("")#blank spaces
tfidf_matrix=tfidf.fit_transform(df_reviews['rating_category'])#create matrix
```

```python
cosine_sim=linear_kernel(tfidf_matrix, tfidf_matrix) #here well get indices but we need actual source name name
```

```python
indices=pd.Series(df_reviews.index, index=df_reviews['business_name']).drop_duplicates()
indices
```

```python
indices['Miss Pizza']
```

```python
def get_recommendations(rating_category,cosine_sim = cosine_sim): # this cell contains similar contents to next cell whatever will be displayed
    idx=indices[rating_category]
    sim_scores=enumerate(cosine_sim[idx])
    sim_scores = sorted(sim_scores, key=lambda x: x[1][0])
    sim_scores=sim_scores[1:11]
    sim_index=[i[0] for i in sim_scores]
    print(df_reviews["rating_category"].iloc[sim_index])
    

```

```python
get_recommendations("Popeyes")
```
